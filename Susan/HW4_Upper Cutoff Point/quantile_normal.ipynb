```python
import matplotlib.pyplot as plt
import numpy as np
import pylab

mu=0
sigma=1
number = 100000

np.random.seed(100)
dataset = np.random.normal(mu, sigma, number)

count, bins, ignored = plt.hist(dataset, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=2, color='r')
plt.show()


q25 = np.quantile(dataset, 0.25)
q75 = np.quantile(dataset, 0.75)

print("\nQuantile 0.25: ", q25)
print("\nQuantile 0.75: ", q75)

w = q75 + 1.5*(q75-q25)
print("\nUpper Outliner Identication: ",w)
```

```python
%matplotlib inline

# Make the figures big enough for the optically challenged.
pylab.rcParams['figure.figsize'] = (10.0, 8.0)

#convert data frame to array for plotting
plt.boxplot(dataset)
plt.ylabel(("Values"))
plt.title("Normal Distribution")
```

```python
from scipy import stats

# Calculate cdf with x=w
phi = stats.norm.cdf(w, loc = 0, scale = 1)
print("cdf of w is: ", phi)

# Calculate probablity for outside the cutpoint of w
p = 2*(1-phi)
print("probability of data out side the upper cut point = ", p)
```
