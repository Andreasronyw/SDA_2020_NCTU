```python
import matplotlib.pyplot as plt
import numpy as np

import sys
from descartes import PolygonPatch
import alphashape
```

```python
def dist(R=4., width=1., num=1000, uniform=True):
    r = np.random.normal(R, width, num)
    phi = np.linspace(0,2.*np.pi, len(r))
    x= r * np.sin(phi)
    y = r* np.cos(phi)
    return x,y
```

```python
x1,y1 = dist(4,0.4)
x2,y2 = dist(0,0.6)
x=np.concatenate((x1,x2),axis=0)
y=np.concatenate((y1,y2),axis=0)

dataset= [i for i in zip(x, y)]
```

```python
q25 = np.quantile(dataset, 0.25)
q75 = np.quantile(dataset, 0.75)

print("\nQuantile 0.25: ", q25)
print("\nQuantile 0.75: ", q75)

w = q75 + 1.5*(q75-q25)
print("\nUpper Outliner Identication: ",w)
```

```python
alpha_shape = alphashape.alphashape(dataset)

fig, ax = plt.subplots()
ax.scatter(*zip(*dataset))
ax.add_patch(PolygonPatch(alpha_shape, alpha=w))
plt.show()
```
